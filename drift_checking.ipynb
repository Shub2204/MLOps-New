```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import load_boston
import warnings
warnings.filterwarnings("ignore")

dic1 = load_boston()
dic1

dic1.target

df1 = pd.DataFrame(dic1.data, columns=dic1.feature_names)
df1.head()
y = pd.DataFrame(dic1.target,columns=['target'])
print(df1.head())
print(df1['CRIM'].min(),df1['CRIM'].max())
print(df1['ZN'].min(),df1['ZN'].max())
print(df1['INDUS'].min(),df1['INDUS'].max())
print(df1['CHAS'].min(),df1['CHAS'].max())
print(df1['NOX'].min(),df1['NOX'].max())
print(df1['RM'].min(),df1['RM'].max())
print(df1['AGE'].min(),df1['AGE'].max())
print(df1['DIS'].min(),df1['DIS'].max())
print(df1['RAD'].min(),df1['RAD'].max())
print(df1['TAX'].min(),df1['TAX'].max())
print(df1['PTRATIO'].min(),df1['PTRATIO'].max())
print(df1['B'].min(),df1['B'].max())
print(df1['LSTAT'].min(),df1['LSTAT'].max())
len(df1)


df2 = pd.DataFrame(columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'])
df2['CRIM']=np.linspace(0.00632,88.9762,100)
df2['ZN']=np.linspace(0.0,100.0,100)
df2['INDUS']=np.linspace(0.46,27.74,100)
df2['CHAS']=np.linspace(0.0,1.0,100)
df2['NOX']=np.linspace(0.385,0.871,100)
df2['RM']=np.linspace(3.561,8.78,100)
df2['AGE']=np.linspace(2.9,100.0,100)
df2['DIS']=np.linspace(1.1296,12.1265,100)
df2['RAD']=np.linspace(1.0,24.0,100)
df2['TAX']=np.linspace(187.0,711.0,100)
df2['PTRATIO']=np.linspace(12.6,22.0,100)
df2['B']=np.linspace(0.32,396.9,100)
df2['LSTAT']=np.linspace(1.73,37.97,100)
df2
```

## Checking for the data drift

```python
from scipy import stats
p_value = 0.05
rejected = 0
for col in df2.columns:
    test = stats.ks_2samp(df1[col], df1[col])
    if test[1] < p_value:
        rejected +=1
        print("Column rejected", col)
        
print(f"We have rejected {rejected} columns in total")
```

There was no data drift observed

## Checking for the Target Drift

```python
import pickle
loaded_model = pickle.load(open('boston_house.pkl','rb'))
y_predict = loaded_model.predict(df2.values)
print(y_predict)
```

```python
sns.distplot(y);
```

```python
sns.distplot(y_predict);
```

We can see that the distribution of target has changed and hence we can conclude that there exists target drift

Since target drift exists we can not calculate concept drift.
